```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

lam = 2  
size = 1000  

data = random.poisson(lam=lam, size=size)

sns.displot(data, kde=False, bins=np.arange(-0.5, max(data)+1.5, 1), color='skyblue', edgecolor='black')

plt.title(f"Poisson Distribution (λ={lam})")
plt.xlabel("Number of Events")
plt.ylabel("Frequency")
plt.grid(True)

plt.show()
```

```python
m_tot = 200
m= np.array(range(1, m_tot + 1))
z_i = 0.5
z_f = 0.1

plt.plot(m, z_i * np.exp( - m * np.log(z_i / z_f) / (m_tot)), label='Exponential Decay', color='orange')
```

```python
# Define the sigmoid function
def sigmoid(x, L, k, x0):
    """
    Sigmoid function.
    
    Parameters
    ----------
    x : float
        Input value.
    L : float
        Maximum value of the function.
    k : float
        Steepness of the function.
    x0 : float
        Midpoint of the function.
        
    Returns
    -------
    float
        Value of the sigmoid function at x.
    """
    return L / (1 + np.exp(k*(x0-x)))
```

```python
# Run the sigmoid function with some parameters
x = np.linspace(0, 2, 100)
L = 1
k = 10
x0 = 1
y = sigmoid(x, L, k, x0)
y1 = sigmoid(x, 1, 0.1, 1)
```

```python
# Plot the sigmoid function
plt.plot(x, y)
plt.plot(x, y1, 'r--')
plt.title("Sigmoid function")
plt.grid()
plt.show()
```

```python
# Basis data
costs = [20, 30, 25]
costs = np.array(costs)
comps = [25/20, 20/30, 20/25]
comps = np.array(comps)
market_shares = [1/5, 1/3, 1-8/15]
market_shares = np.array(market_shares)
```

```python
# Hyperparameters
k1 = 10
k2 = k1 * 0.667
```

```python
# My method
sigmoids = sigmoid(comps, 1, k1, 1)
invest_shares1 = sigmoids * market_shares
invest_shares1 = invest_shares1 / np.sum(invest_shares1)
```

```python
# CEKs method
alpha = comps + market_shares # How is this constructed? No
invest_shares2 = np.exp(k2*alpha)
invest_shares2 = invest_shares2 / np.sum(invest_shares2)
```

```python
# Plot the results
plt.plot(costs, invest_shares1, 'o', label='My method')
plt.plot(costs, invest_shares2, 'o', label='CEKs method')
plt.title("Investment shares")
plt.xlabel("Costs")
plt.ylabel("Investment shares")
plt.legend()
```
